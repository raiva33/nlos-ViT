```python
import pandas as pd
import matplotlib.pyplot as plt

# Set the global font size
font_size = 14
plt.rcParams.update({'font.size': font_size})

# Reading the training loss from the CSV file
training_data = pd.read_csv("/home/mao/Documents/code/mitsuba2-transient-nlos/test_final5_hole.csv")
training_loss = training_data['Value']
training_step = training_data["Step"]

# Reading the validation loss from the CSV file
validation_data = pd.read_csv('/home/mao/Documents/code/mitsuba2-transient-nlos/test_final5_hole_val.csv')
validation_loss = validation_data['Value']
validation_step = validation_data["Step"]

# Plotting both the training and validation loss
plt.figure(figsize=(10, 6))
plt.plot(training_step,training_loss, label='Training Loss', color='blue')
plt.plot(validation_step, validation_loss, label='Validation Loss', color='red')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```
