```python
import os

def compare_directories(directory1, directory2):
    # Remove file extensions from directory1 files
    files_directory1 = set(
        os.path.splitext(filename)[0].replace('_target', '')
        for filename in os.listdir(directory1)
    )

    # Remove file extensions from directory2 files and "_target" string
    files_directory2 = set(
        os.path.splitext(filename)[0].replace('_target', '')
        for filename in os.listdir(directory2)
    )

    files_only_in_directory1 = files_directory1 - files_directory2

    return files_only_in_directory1


# Example usage
directory1 = './train/target/Z'
directory2 = './train/feature/Z'

missing_files = compare_directories(directory1, directory2)
i=0
print(f"Files in {directory1} that are not in {directory2}:")
for file in missing_files:
    print(file)
    i=i+1
print(i)
```

```python
import os
from collections import defaultdict

def find_repeated_files(directory):
    file_dict = defaultdict(list)

    # Traverse through all files in the directory
    for root, dirs, files in os.walk(directory):
        for filename in files:
            full_path = os.path.join(root, filename)

            # Store file paths by their size
            file_size = os.path.getsize(full_path)
            file_dict[file_size].append(full_path)

    repeated_files = []
    for file_size, file_paths in file_dict.items():
        if len(file_paths) > 1:
            repeated_files.extend(file_paths)

    return repeated_files

# Example usage
directory = './train/target_jpg/Z'

repeated_files = find_repeated_files(directory)
i=0
print("Repeated files within the directory:")
for file in repeated_files:
    print(file)
    i=i+1
print(i)
```
