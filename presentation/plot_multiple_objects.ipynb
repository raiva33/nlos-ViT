```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random

# Specify the directory where all the folders are located
parent_directory = '/home/mao/Documents/code/mitsuba2-transient-nlos/predict/predict_final_hole_2batch_extended'

# List all the folders inside the parent directory
folders = [f for f in os.listdir(parent_directory) if os.path.isdir(os.path.join(parent_directory, f))]
print(folders)
# Set up a figure with enough rows for all folders and 4 columns
fig, axs = plt.subplots(len(folders), 4, figsize=(20, 5 * len(folders)))

for row, folder in enumerate(folders):
    # List all images inside the folder
    images = [img for img in os.listdir(os.path.join(parent_directory, folder)) if img.endswith('.png')] 
    print(images)
    # Pick 4 random images
    random_images = random.sample(images, 4)

    # Plot the randomly selected images
    for col, image in enumerate(random_images):
        img_path = os.path.join(parent_directory, folder, image)
        img = mpimg.imread(img_path)
        axs[row, col].imshow(img, cmap="gray")
        axs[row, col].axis('off') # Remove axes
        axs[row, col].text(0.5, -0.1, image[:-10], size=19, ha="center", transform=axs[row, col].transAxes) # Add the filename as a label below the image


plt.show()
```
